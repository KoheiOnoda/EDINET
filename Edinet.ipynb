```python
import requests
import json
import zipfile
import os

url = "https://disclosure.edinet-fsa.go.jp/api/v1/documents.json"
params = {
    "date" : '2019-04-25',
    "type" : 2
}
res = requests.get(url, params=params)
#print(res.text)

res_text = json.loads(res.text)
results = res_text["results"]
kessan = []
for result in results:
    if result['docDescription'] is not None:
        if '四半期' in result['docDescription']:
            print(result['docID'], result['docDescription'], result['filerName'])
            kessan.append(result)
            
for result in kessan:
    docid = result['docID']
    url = "https://disclosure.edinet-fsa.go.jp/api/v1/documents/" + docid
    params = { "type" : 1 }
    res = requests.get(url, params=params)
    foldername = "kessan/" + docid
    filename = foldername + ".zip"
    if res.status_code == 200:
        if(os.path.exists(filename) == False):
            with open(filename, 'wb') as f:
                for chunk in res.iter_content(chunk_size=1024):
                    f.write(chunk)
            with zipfile.ZipFile(filename) as existing_zip:
                existing_zip.extractall(foldername)

import glob
from bs4 import BeautifulSoup
for result in kessan:
    docid = result['docID']
    filepath = 'kessan/' + docid + '/XBRL/PublicDoc/'
    files = glob.glob(filepath+'*.htm')
    files = sorted(files)
    target_file = files[1]
    with open(target_file , encoding='utf-8') as f:
        html = f.read()
        soup = BeautifulSoup(html, 'html.parser')
        tag_p = soup.find_all('p')
        for p in tag_p:
            print(p.text)
```
